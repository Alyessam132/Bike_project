#### filter data

```python
CITY_DATA = { 'chicago': 'chicago.csv',
              'new york city': 'new_york_city.csv',
              'washington': 'washington.csv' }

time_filter_list = ['month', 'day', 'both', 'none']
month_list = ['january', 'february', 'march', 'april', 'may', 'june']
day_list = ['saturday', 'sunday', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday']
time_filter = ''
city = ''

def get_filters():
    
    '''
    Asks user to specify a city, month, and day to analyze.

    Returns:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or 'all' to apply no month filter
        (str) day - name of the day of week to filter by, or 'all' to apply no day filter
    '''
    
    global city
    global time_filter
    
    print('Hello! Let\'s explore some US bikeshare data!')
    # TO DO: get user input for city (chicago, new york city, washington). HINT: Use a while loop to handle invalid inputs
    city = input('Would you like to see data for Chicago, New York City, or Washington?\n').lower()
    
    while city not in CITY_DATA.keys() :
        print('Invalid Value')
        city = input('Would you like to see data for Chicago, New York City, or Washington?\n').lower()

   
    time_filter = input('Would you like to filter the data by month, day, both, or not at all?\
    Type \'None\' for no time filter\n').lower()
    
    while time_filter not in time_filter_list :
        print('Invalid Value')
        time_filter = input('Would you like to filter the data by month, day, both, or not at all?\
    Type \'None\' for no time filter\n').lower()
        
    # TO DO: get user input for month (all, january, february, ... , june)
    if (time_filter == 'month') or  (time_filter == 'both') :
        month = input('Which Month? January, February, March, April, May, or June?\
        Please, type out the full month name\n').lower()
        
        while month not in month_list :
            print('Invalid Value')
            month = input('Which Month? January, February, March, April, May, or June?\
        Please, type out the full month name\n').lower()
            
    else :
        month = 'all'
    
    
    # TO DO: get user input for day of week (all, monday, tuesday, ... sunday)
    if (time_filter == 'day') or  (time_filter == 'both') :
        day = input('Which Day?  Please, type out the full day name\n').lower()
        
        while day not in day_list :
            print('Invalid Value')
            day = input('Which Day?  Please, type out the full day name\n').lower()
            
    else :
        day = 'all'

    print('-'*40)
    return city, month, day
```

#### Load Data Function

```python
def load_data(city, month, day):
    
    '''
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or 'all' to apply no month filter
        (str) day - name of the day of week to filter by, or 'all' to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    '''
    df = pd.read_csv(CITY_DATA[city])
    df['Start Time'] = df['Start Time'].astype('datetime64')
    df['Month'] = df['Start Time'].dt.month_name()
    df['Week Day'] = df['Start Time'].dt.day_name()
    
    
    if month != 'all' :
        df = df[df['Month'] == month.title()]
    
    if day != 'all' :
        df = df[df['Week Day'] == day.title()]

    return df
```

#### TIme Stats

```python
def time_stats(df):
    """Displays statistics on the most frequent times of travel."""
    
    global time_filter
    
    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()

    # TO DO: display the most common month
    month_count = df['Month'].value_counts()
    month_count_max = month_count.max()
    month_most_common = month_count.idxmax()
    
    if (time_filter == 'day') or (time_filter == 'none') :
        print('\nCalculating The Most Frequent Month of Travel...\n')
        print('Most Common Month: {}, Count: {}, Filter: {}'.format(month_most_common,month_count_max,time_filter))



    # TO DO: display the most common day of week
    week_day_count = df['Week Day'].value_counts()
    week_day_count_max = week_day_count.max()
    week_day_most_common = week_day_count.idxmax()
    
    if (time_filter == 'month') or (time_filter == 'none') :
        print('\nCalculating The Most Frequent Day of Travel...\n')
        print('Most Common Day: {}, Count: {}, Filter: {}'.format(week_day_most_common,week_day_count_max,time_filter))

    # TO DO: display the most common start hour
    df['Start Hour'] = df['Start Time'].dt.hour
    start_hour_count = df['Start Hour'].value_counts()
    start_hour_count_max = start_hour_count.max()
    start_hour_most_common = start_hour_count.idxmax()
    
    print('\nCalculating The Most Frequent Hour of Travel...\n')
    print('Most Common Start Hour: {}, Count: {}, Filter: {}'.format(start_hour_most_common,start_hour_count_max,time_filter))

    print('\nThis took %s seconds.' % (time.time() - start_time))
    print('-'*40)
```

#### Station Stats

```python
def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # TO DO: display most commonly used start station
    start_station_count = df['Start Station'].value_counts()
    start_station_count_max = start_station_count.max()
    start_station_most_common = start_station_count.idxmax()
    
    print('\nCalculating The Most Start Station...\n')
    print('Most Common Start Station: {}, Count: {}'.format(start_station_most_common,start_station_count_max))
    
    # TO DO: display most commonly used end station
    end_station_count = df['End Station'].value_counts()
    end_station_count_max = end_station_count.max()
    end_station_most_common = end_station_count.idxmax()
    
    print('\nCalculating The Most Popular End Station...\n')
    print('Most Common End Station: {}, Count: {}'.format(end_station_most_common,end_station_count_max))

    # TO DO: display most frequent combination of start station and end station trip
    df['Trip'] = df['Start Station'] + ', ' + df['End Station']
    trip_count = df['Trip'].value_counts()
    trip_count_max = trip_count.max()
    trip_most_common = trip_count.idxmax()
    
    print('\nCalculating The Most Popular Trip...\n')
    print('Most Common Trip: {}, Count: {}'.format(trip_most_common,trip_count_max))

    print('\nThis took %s seconds.' % (time.time() - start_time))
    print('-'*40)
```

#### Trip duration Stats

```python
def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""

    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # TO DO: display total travel time
    total_travel_time = df['Trip Duration'].sum()
    
    print('\nCalculating Total Trip Duration...\n')
    print('Total Travel Time: {}'.format(total_travel_time))

    # TO DO: display mean travel time
    print('\nCalculating Average Trip Duration...\n')
    average_travel_time = df['Trip Duration'].mean()
    print('Average Travel Time: {}'.format(average_travel_time))

    print('\nThis took %s seconds.' % (time.time() - start_time))
    print('-'*40)
```

#### User Stats

```python
def user_stats(df):
    """Displays statistics on bikeshare users."""
    global city
    
    print('\nCalculating User Stats...\n')
    start_time = time.time()

    # TO DO: Display counts of user types
    
    print('\nCalculating User Types...\n')
    type_count = df['User Type'].value_counts()
    
    for i in range(len(type_count)) : 
        print ('{}: {}'.format(type_count.index[i],type_count[i]))


    # TO DO: Display counts of gender
    if city != 'washington' :
        
        print('\nCalculating User Gender...\n')
        gender_count = df['Gender'].value_counts()
        
        for i in range(len(gender_count)) : 
            print ('{}: {}'.format(gender_count.index[i],gender_count[i]))
        

    # TO DO: Display earliest, most recent, and most common year of birth
    if city != 'washington' :
        
        print('\nCalculating User Year of birth...\n')
        earliest_year = df['Birth Year'].min()
        most_recent_year = df['Birth Year'].max()
        most_common_year = df['Birth Year'].value_counts().idxmax()
        
        print('Earliest Year: {}'.format(earliest_year))
        print('Most Recent Year: {}'.format(most_recent_year))
        print('Most Common Year: {}'.format(most_common_year))
        

    print('\nThis took %s seconds.' % (time.time() - start_time))
    print('-'*40)
```

#### Display User Data

```python
def display_user_data():
    """Displays data of five random users from the selected city and asking for input to display more random users."""
    global city
    df = pd.read_csv(CITY_DATA[city])
    df.rename(columns={'Unnamed: 0':'',}, inplace=True)
    df.fillna("Not Determined", inplace=True)
    
    df_random = df.sample(5).to_dict(orient='records')
    
    while True :
        
        print('\nDisplaying random individual trip data...\n')
        
        for item in df_random :
            print(json.dumps(item,indent =1 ))
            
        more_users = input('\nWould you like to view more individual trip data? Enter yes or no.\n').lower()
        
        if more_users != 'yes' :
            break
    
    
    
```

#### Main Function

```python
def main():
    while True:
        
        city, month, day = get_filters()
        df = load_data(city, month, day)
        print(df)        
        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df)
        display_user_data()

        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break


if __name__ == '__main__':
        main()
        
```
